Experiment 2:

  a.  Design a single unit perceptron for classification of a linearly separable
binary dataset without using pre-defined models. Use the Perceptron() from
sklearn.

 b.  Identify the problem with single unit Perceptron. Classify using Or, And and Xor-ed data and analyze the result.



---





```python
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score

X = np.array([
    [1, 1],
    [2, 1],
    [1, 2],
    [0, 0],
    [0, 1],
    [1, 0]
])

y = np.array([1, 1, 1, 0, 0, 0])

model = Perceptron(max_iter=1000, eta0=0.1, random_state=42)

model.fit(X, y)

y_pred = model.predict(X)

accuracy = accuracy_score(y, y_pred)

print("Weights:", model.coef_)
print("Bias:", model.intercept_)
print("Predictions:", y_pred)
print("Accuracy:", accuracy)
```

```python
X_and = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y_and = np.array([0, 0, 0, 1])

model_and = Perceptron(max_iter=1000, eta0=0.1)
model_and.fit(X_and, y_and)

y_pred_and = model_and.predict(X_and)

print("AND Predictions:", y_pred_and)
```

```python
X_or = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y_or = np.array([0, 1, 1, 1])

model_or = Perceptron(max_iter=1000, eta0=0.1)
model_or.fit(X_or, y_or)

y_pred_or = model_or.predict(X_or)

print("OR Predictions:", y_pred_or)
```

```python
X_xor = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y_xor = np.array([0, 1, 1, 0])

model_xor = Perceptron(max_iter=1000, eta0=0.1)
model_xor.fit(X_xor, y_xor)

y_pred_xor = model_xor.predict(X_xor)

print("XOR Predictions:", y_pred_xor)
```
